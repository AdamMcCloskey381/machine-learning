```python
from sklearn.impute import SimpleImputer
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
import pandas as pd 
import numpy as np
```

```python
iris = load_iris()
df = pd.DataFrame(data = np.c_[iris["data"], iris["target"]],
                  columns = iris.feature_names + ["target"])
df_raw = df.copy()
df.head()
```

```python
Sc = StandardScaler()
for col in df.columns[:len(df.columns.tolist())-1].tolist():
  df[col] = (df[col] - df[col].mean())/df[col].std()
```

```python
df.describe().T
```

```python
df_raw.describe().T
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree1 = DecisionTreeClassifier()
x = df.loc[:, df.columns != "target"]
y = df.loc[:, df.columns == "target"]
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 123)
tree1.fit(x_train, y_train)
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, tree1.predict(x_test))
score
```

```python
tree2 = DecisionTreeClassifier()
x1 = df_raw.loc[:, df_raw.columns != "target"]
y1 = df_raw.loc[:, df_raw.columns == "target"]
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.25, random_state = 123)
tree2.fit(x1_train, y1_train)
score = accuracy_score(y1_test, tree2.predict(x1_test))
score
```
