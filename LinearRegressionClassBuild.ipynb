```python
class LinearR:
  
  def __init__(self, coef = [], intercept = []):
    self.coef = coef
    self.intercept = intercept
  
  def fit(self, X, Y, n_iter = 3000, eta = 0.01):
    Y = Y.reshape(-1, 1)
    m = len(X)
    bias = np.ones((m, 1))
    X = np.c_[bias, X]
    weights = np.random.randn(2, 1)
    self.eta = eta
    self.n_iter = n_iter
    
    for i in range(3000):
      gradient = (2/m) * X.T.dot(X.dot(weights) - Y)
      weights = weights - self.eta * gradient
      self.intercept.append(weights[0][0])
      self.coef.append(weights[1][0])
    self.coef = self.coef[-1]
    self.intercept = self.intercept[-1]
    
  def predict(self, X):
    self.pred = []
    for element in X:
      self.pred.append(float(self.intercept + self.coef * element))
    
```

```python
lr = LinearR()
lr.fit(X = data, Y = target)
lr.intercept
```

```python
plt.figure(figsize = (12, 12))
plt.title("linear regression")
plt.xlabel("x variable")
plt.ylabel("y variable")
plt.scatter(data, target, label = "data points", c = "gold", alpha = 0.5)
plt.plot(data, lr.intercept + lr.coef*data, label = "model", c = "red", alpha = 0.7)
```
